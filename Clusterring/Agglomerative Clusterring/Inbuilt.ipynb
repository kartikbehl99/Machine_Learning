# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy as sch
import pandas as pd
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D, proj3d
```

# Loading sample data

```python
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

# Making the dendrogram

```python
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward')) #The ward method tries to minimise the variance of the data
#There are other methods as well
plt.title('Dendrograms')
plt.xlabel('')
plt.ylabel('Euclidean distances')
plt.Figure(figsize(20,10))
plt.show()
```

# Using agglomerative clustering using the dendrogram obtained

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
#We are setting n_clusters to 3 because in the above denndrogram, the 3 in middle are with minimum difference between them
#affinity to refers to the distance metric to be used
y_hc = hc.fit_predict(X)
```

# Plotting the clusters obtained

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(X[:,0], X[:,1], X[:,2], c = y_hc)
# plt.scatter(X[:,0],X[:,1], c=y_hc)
# plt.scatter(X[:,0],X[:,2], c=y_hc)
# plt.scatter(X[:,0],X[:,3], c=y_hc)
# # plt.scatter(X[:,0],Y, c=y_hc)
# plt.show()
```
